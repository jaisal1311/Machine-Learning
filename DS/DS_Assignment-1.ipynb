```python
Y = np.random.rand(10,10)
print(Y)

#a
detMat = np.linalg.det(Y)
print(detMat)

#b
rankMat = np.linalg.matrix_rank(Y)
print(rankMat)

#c
diagMat = np.diagonal(Y)
print(diagMat)

#d
traceMat = np.trace(Y)
print(traceMat)

#e
eigValMat = np.linalg.eigvals(Y)
print(eigValMat)

#f
svdMat = np.linalg.svd(Y)
print(svdMat)

#g
if(np.linalg.det(Y) != 0):
    invMat = np.linalg.inv(Y)
    print('Exists')
    print(invMat)
else:
    print('Not Exists')
    print(np.linalg.pinv(Y))
    
#h
mulMat = np.multiply(Y, np.linalg.inv(Y) if np.linalg.det(Y) != 0 else np.linalg.pinv(Y))
print(mulMat)
    
```
